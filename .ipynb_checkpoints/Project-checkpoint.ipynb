```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn import linear_model
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error
from sklearn.preprocessing import StandardScaler
from sklearn.feature_selection import RFE
import math
from sklearn.model_selection import validation_curve
from sklearn.metrics import mean_absolute_error
from sklearn.model_selection import StratifiedKFold
from sklearn.feature_selection import RFECV
import matplotlib.dates as mdates
import datetime
from pandas.api.types import CategoricalDtype


import seaborn as sns
import matplotlib.pylab as plt
%matplotlib inline
```

```python
#Combining Data
testingdata = pd.read_csv("data/testing.csv")
trainingdata = pd.read_csv("data/training.csv")
newdata = testingdata.append(trainingdata)
newdata = newdata.dropna()
newdata.head()

#Creating Hour Column
newdata["Hour"] = pd.DatetimeIndex(newdata["date"]).hour
newdata.head()

#Creating Date column
newdata["date2"] = pd.DatetimeIndex(newdata["date"].iloc[0:19735])

#Converting WeekStatus to 0 and 1
newdata["WeekStatus"] = newdata.WeekStatus.replace(to_replace = ["Weekday", "Weekend"], value = [0, 1])
```

## Pair Plot

```python
#Pair Plot

#Correlation Function

def correlationplot(x , y, **kws):
    corrmat = np.corrcoef(x, y)
    pearson = round(corrmat[0, 1], 2)
    ax = plt.gca()
    ax.annotate(pearson, [.5, .5], xycoords = "axes fraction", ha = "center", va = "center", fontsize = 30)
    
#Plot    
    
pairplot = sns.PairGrid(newdata, vars = ["Appliances", "lights", "RH_1", "T2", "RH_2", "T3", "RH_3"], x_vars = ["Appliances"], y_vars = ["lights", "RH_1", "T2", "RH_2", "T3", "RH_3"])
pairplot.map_upper(correlationplot)
pairplot.map_lower(sns.regplot, color = "black", scatter_kws = {"s":3}, line_kws = { "color" : "r"})
pairplot.map_diag(plt.hist, color = "lightblue")
plt.show()

#The graph was reproduced from the research paper and shows the relationship between Appliances and lights, RH1, T2, RH2, T3 and RH3.
#Some interesting correlations are lights and Appliances which has a 0.20 correlation, RH_1 and RH_2 which have a 0.80 correlation, RH_1 and RH_3 which have a 0.84 correlation and T2 and T3 which has a 0.74 correlation.
```

## Linear Regression Set Up

```python
#Linear Regression Model

#Declaring variables

X = newdata[["lights","T1","RH_1","T2","RH_2","T3","RH_3","T4","RH_4","T5","RH_5","T6","RH_6","T7","RH_7","T8","RH_8","T9","RH_9","T_out","Press_mm_hg","RH_out","Windspeed","Visibility","Tdewpoint","rv1","rv2","NSM"]]
y = newdata["Appliances"]
```

## Linear Regression with non-numeric data included

```python
#Dummy Variable Conversion for Day_of_week

set(newdata["Day_of_week"])
#set(newdata["WeekStatus"])

day_of_week_df = pd.get_dummies(newdata["Day_of_week"])
#WeekStatus_df = pd.get_dummies(newdata["WeekStatus"])
#WeekStatus_df
combined_df = newdata
days_df = day_of_week_df.rename(columns={"M":"Monday", "Tu": "Tuesday", "W": "Wednesday", "Th": "Thursday", "F": "Friday", "Sa": "Saturday", "Su": "Sunday"})
final_df = combined_df.join(days_df)
final_df
```

```python
#Modelling with dummy variables
train, test = train_test_split(final_df, test_size = 0.25, random_state = 142)
X_train = train.drop(["date", "date2", "Hour", "Appliances", "Day_of_week"], axis = 1)
y_train = train["Appliances"]
X_test = test.drop(["date", "date2", "Hour", "Appliances", "Day_of_week"], axis = 1)
y_test = test["Appliances"]

#Initiating model
reg2 = linear_model.LinearRegression()

#Model Training
reg2.fit(X_train, y_train)

#Predicting on training and testing data
y_train_hat = reg2.predict(X_train)
y_test_hat = reg2.predict(X_test)

#Evaluation
print("Parameters ", ["lights","T1","RH_1","T2","RH_2","T3","RH_3","T4","RH_4","T5","RH_5","T6","RH_6","T7","RH_7","T8","RH_8","T9","RH_9","T_out","Press_mm_hg","RH_out","Windspeed","Visibility","Tdewpoint","rv1","rv2","NSM", "WeekStatus"])
print()
print("Scores on training set: ")
print("RMSE: ", math.sqrt(mean_squared_error(y_train, y_train_hat)))
print("R2: ", r2_score(y_train, y_train_hat))
print("MAE: ", mean_absolute_error(y_train, y_train_hat))
print()
print("Scores on testing set: ")
print("RMSE: ", math.sqrt(mean_squared_error(y_test, y_test_hat)))
print("R2: ", r2_score(y_test, y_test_hat))
print("MAE: ", mean_absolute_error(y_test, y_test_hat))
#These results are more appropriate considering they use all variables required.

#With the RMSE (Root Mean Square Error) value in the test set being larger than the RMSE value in the training set, it indicates that the model performs testing well with the sample data but will have little predictive value outside of the sample data.

#The R^2 score for the training set is 0.17 whilst the testing set is 0.16. With these low values, it indicates that there's low variability. Indicating that the model doesn't fit the data well.

#The MAE (Mean Absolute Error) score for the training set is 53.53 whilst the MAE score for the testing set is 54.32. This indicates there are a significant amount of errors between the grouped observations.
```

## Feature Importance

```python
#RFECV Feature Importance
regest = linear_model.LinearRegression()
k = StratifiedKFold(n_splits = 10, random_state = 42)
rfecv = RFECV(estimator = regest, step = 1, cv = k, scoring = "neg_mean_squared_error")
rfecv = rfecv.fit(X_train, y_train)

print("Optimal number of features :", rfecv.n_features_)
print("Best features :", X_train.columns[rfecv.support_])

#Plot
plt.figure(figsize = (20,10))
plt.xlabel("Variables")
plt.ylabel("RMSE (Cross-Validation)")
plt.title("RMSE scores on test set using RFECV")

rmse_cv_scores = np.sqrt(-rfecv.grid_scores_)
p = plt.plot(range(1, len(rfecv.grid_scores_) + 1), rmse_cv_scores, marker = "o", color = "lightblue")

#From the RFECV the optimal number of features is 36, whilst the best features can be seen within the print.
#A visualisation of the feature importance can be seen in the plot.

#Compared to the research paper, the results are similar, as the optimal number of features within the paper was 35 and the number produced in the portfolio is 36.
```
